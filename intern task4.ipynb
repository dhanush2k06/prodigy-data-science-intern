```python
col_names=["ID","Entity","Sentiments","Contest"]
data=pd.read_csv("twitter_training.csv", names=col_names)
data
```

```python
sentiment_count=data["Sentiments"].value_counts()
sentiment_count
```

```python
plt.figure(figsize=(6,3))
sentiment_count.plot(kind="bar",color=['red','green','blue','yellow'])
plt.title("SEntiment Distribution")
plt.xlabel("Number of tweets")
plt.xticks(rotation=0)
plt.show()
```

```python
brand_data=data[data["Entity"].str.contains("Microsoft",case=False)]
brand_sentiment_count=brand_data['Sentiments'].value_counts()
brand_sentiment_count
```

```python
plt.figure(figsize=(6,6))
plt.pie(brand_sentiment_count,labels=brand_sentiment_count.index,autopct='%1.11f%%', startangle=140)
plt.show()
```
